```python
def get_final_cost_weight(weight,cost, answer_indexes):
  cost_f = 0
  weight_f = 0
  for i in range(len(answer_indexes)):
    if answer_indexes[i] == 1:
      cost_f += cost[i]
      weight_f += weight[i]
  return cost_f, weight_f
```

```python
import time
import numpy as np
def check_time(f, *args):
  function_time = 0.
  for i in np.arange(1000):
    start = time.time()
    f(*args)
    function_time += time.time() - start
  avarege_time = function_time / 1000
  return avarege_time
```

## 2-approx algorithm

```python
def greed(max_weight, weight, cost, count):
    sorted_indices = sorted(range(len(cost)), key=lambda k: cost[k])
    object_id = [0 for _ in range(len(cost))]
    rest_capacity = max_weight
    cost_ = 0
    for i in reversed(sorted_indices):
        if weight[i] <= rest_capacity:
            count += 1
            cost_ += cost[i]
            object_id[i] = 1
            rest_capacity -= weight[i]
    return cost_, object_id, count

def pro_greed(max_weight, weight, cost, count):
    quality = [round(cost[i]/weight[i]) for i in range(len(cost))]
    sorted_indices = sorted(range(len(quality)), key=lambda k: quality[k])
    object_id = [0 for _ in range(len(cost))]
    rest_capacity = max_weight
    cost_ = 0
    for i in reversed(sorted_indices):
        if weight[i] <= rest_capacity:
            count += 1
            cost_ += cost[i]
            object_id[i] = 1
            rest_capacity -= weight[i]
    return cost_, object_id, count


def two_approx(max_weight, weight, cost):
    count = 0
    answ_cost, answ_objects,count_1 = greed(max_weight, weight, cost, count)
    pro_answ_cost, pro_answ_objects,count_2 = pro_greed(max_weight, weight, cost, count)
    count = count_1 + count_2
    if answ_cost > pro_answ_cost:
        return answ_objects, count
    else:
        return pro_answ_objects, count
```

## Dynamic programming on the scales

```python
def table_calculation(max_weight, weight, cost, count):
      n = len(cost)
      V = [[0 for a in range(max_weight + 1)] for i in range(n + 1)]
      for i in range(n+1):
            for a in range(max_weight + 1):
                  if i == 0 or a == 0:
                        V[i][a] = 0
                  elif weight[i-1] <= a:
                        count += 1
                        V[i][a] = max(cost[i-1] + V[i-1][a-weight[i-1]], V[i-1][a])
                  else:
                        V[i][a] = V[i-1][a]
      return V, weight, cost, count

def DP_weight(max_weight, weight, cost):
      count = 0
      V, area, value,count = table_calculation(max_weight, weight, cost, count)
      n = len(value)
      res = V[n][max_weight]
      a = max_weight
      items_list = []
      for i in range(n, 0, -1):
            if res < 0:
                  break
            if V[i-1][a] == res:
                  items_list.append(0)
                  continue
            else:
                  items_list.append(1)
                  res -= value[i-1]
                  a -= area[i-1]
      items_list = items_list[::-1]
      return items_list,count
```

## Branch and bound method using only 1 subject

```python
import queue
from collections import namedtuple

class Node:
    def __init__(self, level, value, weight, bound, contains):
         self.level = level
         self.value = value
         self.weight = weight
         self.bound = bound
         self.contains = contains

def upper_bound(u, k, n, v, w):
        if u.weight > k:
            return 0
        else:
            bound = u.value
            wt = u.weight
            j = u.level
            while j < n and wt + w[j] <= k:
                bound += v[j]
                wt += w[j]
                j += 1
            # fill knapsack with fraction of a remaining item
            if j < n:
                bound += (k - wt) * float(v[j])/ w[j]
            #print(bound)
            return bound

class Items:
  def __init__(self,weight,cost):
    self.weight = weight
    self.value = cost

def knapsack(items, capacity):
        item_count = len(items)
        count = 0
        v = [0]*item_count
        w = [0]*item_count
        # sort items by value to weight ratio
        items = sorted(items, key=lambda k: float(k.value)/k.weight, reverse = True)
        for i,item in enumerate(items, 0):
            v[i] = int(item.value)
            w[i] = int(item.weight)
        q = queue.Queue()
        root = Node(0, 0, 0, 0.0,[])
        root.bound = upper_bound(root, capacity, item_count, v, w)
        q.put(root)
        value = 0
        taken = [0]*item_count
        best = set()
        while not q.empty():
            c = q.get()
            if c.bound > value:
                level = c.level+1
            left = Node(level,c.value + v[level-1], c.weight + w[level-1], 0.0, c.contains[:])
            left.bound = upper_bound(left, capacity, item_count, v, w)
            left.contains.append(level)
            if left.weight <= capacity:
                if left.value > value:
                    count += 1
                    value = left.value
                    best = set(left.contains)
                if left.bound > value:
                    q.put(left)
            right = Node(level,c.value, c.weight, 0.0, c.contains[:])
            right.bound = upper_bound(right, capacity, item_count, v, w)
            if right.weight <= capacity:
                if right.value > value:
                    count += 1
                    value = right.value
                    best = set(right.contains)
                if right.bound > value:
                    q.put(right)
        for b in best:
            taken[b-1] = 1
        #value = sum([i*j for (i,j) in zip(v,taken)])
        return taken,count
```

## PTAS

```python
from itertools import chain, combinations
import numpy
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
def gen_M_list(n,k):
  origin = np.arange(n)
  M_list = list(powerset(origin))
  c = 0
  for i in M_list:
    if len(i) > k :
      M_list.pop(c)
    c += 1
  return M_list
```

```python
def sum_weight_M(weight,M):
  M_weight = 0
  for i in list(M):
    M_weight += weight[i]
  return M_weight

def GS(n,max_weight,cost,weight, M):
  z = 0
  X = []
  M_weight = sum_weight_M(weight,M)
  c = max_weight - M_weight
  for j in range(n):
    if j not in M and weight[j] <= c:
      z = z + cost[j]
      c = c - weight[j]
      X.append(j)
  return z, X

def S(n,max_weight,cost,weight,k):
  count = 0
  z = 0
  X = []
  M_list = gen_M_list(n,k)
  for M in M_list:
    if sum_weight_M(weight,list(M)) <= max_weight:
      z_, X_ = GS(n,max_weight,cost, weight, list(M))
      if z_ + sum_weight_M(cost, list(M)) > z:
        z = z_ + sum_weight_M(cost, list(M))
        X.append(X_ + list(M))
        count += 1

  res =[]
  for i in np.arange(len(cost)):
    if i in X[-1]:
      res.append(1)
    else:
      res.append(0)
  return res, count
```
